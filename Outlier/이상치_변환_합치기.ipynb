```python
a = ""

with open("prose.txt", "r") as f:
    a = f.read()

import re
sentences = re.split(r'\[Sentence \d+\]\.\s', a)

# 빈 문자열 제거
sentences = [sentence for sentence in sentences if sentence]

splitted = [re.split(r'\-\>', sentence) for sentence in sentences]
splitted = [re.sub(r'\s*\[산문\]\n', '', translation) for original, translation in splitted]
splitted = [re.sub(r'\n', ' ', translation) for translation in splitted]
splitted
```
