```python
# 'easy_sentence' 열을 string 타입으로
df['easy_sentence'] = df['easy_sentence'].astype(str)

# 정제에 필요한 함수
def RemoveBefore(text, marker):
  parts = text.split(marker, 1)
  return parts[1] if len(parts) > 1 else text

def RemoveAfter(text, marker):
    parts = text.split(marker, 1)
    return parts[0] if len(parts) > 1 else text
```
