```python
def split_sentence(row, split_points):
    sentences = [row['easy_sentence']]
    for split_point in split_points:
        new_sentences = []
        for sentence in sentences:
            parts = sentence.split(split_point)
            if len(parts) > 1:
                # split point가 문장 내에 있을 때만 처리
                new_sentences.append(parts[0] + split_point)
                new_sentences.extend(parts[1:])
            else:
                new_sentences.append(parts[0])
        sentences = new_sentences
    new_rows = []
    for sentence in sentences:
        new_rows.append({'title': row['title'], 'easy_sentence': sentence.strip(), 'quote_count':row['quote_count']})
    return new_rows

# split points 설정
split_points = ['하거늘,', '말하기를,']

new_rows = []
for index, row in df.iterrows():
    if index == 187:  
        new_rows.extend(split_sentence(row, split_points))
    else:
        new_rows.append({'title': row['title'], 'easy_sentence': row['easy_sentence'], 'quote_count':row['quote_count']})

df = pd.DataFrame(new_rows)
```
