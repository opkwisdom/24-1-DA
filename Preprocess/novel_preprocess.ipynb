```python
def split_sentences(passage, tagger):
    """
    문장을 여러 개로 분리하는 함수
    passage: 여러 개의 문장
    tagger: 바른 tagger
    """
    tagged = tagger.tags([passage], auto_split=True)
    
    sentences = []
    m = tagged.msg()
    for s in m.sentences:
        sentences.append(s.text.content)
    return sentences
```

### 한자 -> 한글로 변경하기

```python
def korean_ratio(sentence):
    """
    한글의 비율을 반환하는 함수
    
    Parameters:
        sentence (str): 분석하고자 하는 문자열
        
    Returns:
        float: 한글의 비율 (0 ~ 1)
    """
    # 한글 문자의 개수를 세기
    korean_count = sum(1 for char in sentence if char >= '가' and char <= '힣')
    space_count = sentence.count(' ')
    
    # 전체 문자의 개수
    total_chars = len(sentence) - space_count
    
    # 한글 비율 계산
    ratio = korean_count / total_chars if total_chars != 0 else 0
    
    return ratio
```

```python
import re

def clean_text(text):
    # 한글과 특수문자 외의 모든 문자를 지우는 정규표현식
    #     한글 유니코드 범위: 완성형 한글 (U+AC00-U+D7AF), 한글 자모 (U+1100-U+11FF)
    #     특수문자 범위: 쉼표, 마침표, 느낌표, 물음표, 따옴표, 괄호 등
    pattern = '[^가-힣0-9\s,\.!?\(\)\[\]\'\"“”‘’~-]'

    # 텍스트에 적용
    cleaned_text = re.sub(pattern, '', str(text))
    
    # 괄호 안의 문자열도 같이 제거
    pattern2 = '\([^\(\)]*\)'

    # 텍스트에 적용
    cleaned_text = re.sub(pattern2, '', cleaned_text)

    return cleaned_text

df['cleaned_content'] = df['sentence'].apply(clean_text)
df
```

```python
# 문장 내 한국어 비율이 50% 이상인 행만 남김
df6 = df3[df3["cleaned_content"].apply(lambda x: korean_ratio(x) < 0.5)]
df6 = df6.reset_index()
df6 = df6.drop(columns="index")
df6
df6.to_csv("df_no_author_info_under50%.csv", index=False)
```
