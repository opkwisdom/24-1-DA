```python
# 'sentiment_label'에 따른 'sentiment_score' 값 업데이트
def UpdateScore(row):
    if row['sentiment_label'] in ['중립', '일상적인']:
        return 0
    elif row['sentiment_label'] in ['슬픔', '분노', '불안', '당황', '혐오', '슬픔(우울한)']:
        return row['sentiment_score'] * (-1)
    elif row['sentiment_label'] == '행복':
        return row['sentiment_score']
    return row['sentiment_score']

df['sentiment_score'] = df.apply(UpdateScore, axis=1)
df
```

### 단락 시작 넘버 json 파일 로드

```python
import json
from collections import OrderedDict

with open("parag_split_ids.json", 'r', encoding='utf-8') as file:
    data = json.load(file)

markers = dict(OrderedDict(sorted(data.items())))

for key, value in markers.items():
  print(f"'{key}': {value},")
```

### 화자/청자 별 등장 횟수 카운트, 단락별로 병합

```python
# eval을 안전하게 사용
def safe_eval(item):
    if isinstance(item, str):
        return eval(item)
    return item

# sentiment_label과 sentiment_score를 리스트로 변환
df['sentiment_label'] = df['sentiment_label'].apply(lambda x: [x] if not pd.isna(x) else [])
df['sentiment_score'] = df['sentiment_score'].apply(lambda x: [x] if not pd.isna(x) else [])

# 딕셔너리 'markers'를 기준으로 병합
def MergebyMarkers(df, markers):
    rows_to_keep = []

    # 열 병합 함수들 정의
    def MergeNums(sub_df, col):
        return sub_df[col].sum()

    def MergeStrs(sub_df, col):
        return ' '.join(sub_df[col].dropna())

    def MergeList(sub_df, col):   # 중복 관계없이 단순 append
        merged_list = []
        for sublist in sub_df[col].dropna().apply(safe_eval).tolist():
            merged_list.extend(sublist)
        return merged_list

    def MergeListsUnique(sub_df, col):    # 중복되지 않게 append
        merged_list = []
        for sublist in sub_df[col].dropna().apply(safe_eval).tolist():
            for item in sublist:
                if item not in merged_list:
                    merged_list.append(item)
        return merged_list

    def MergeDicts(sub_df, col):
        merged = []
        for entry in sub_df[col].dropna().apply(safe_eval):
            merged.append(entry)
        return merged

    def MergeAuthor(sub_df):
        if sub_df['author'].isna().any() or (sub_df['author'] == '미상').any():
            return 'NaN'
        return sub_df['author'].iloc[0]

    number_columns = ['quote_count', 'next_quote_count', 'pre_dialogue', 'line_count', 'sentence_length']
    string_columns = ['cleaned_content', 'complex_words_and_annot', 'easy_sentence', 'quote_dup', 'quote_dup_fixed']
    list_columns_unique = ['NE_tagged', 'NE_tagged_set', 'NE_tagged_modernized', 'NE_tagged_modernized_set']
    list_append_columns = ['sentiment_label', 'sentiment_score']

    for title, page_markers in markers.items():
        df_title = df[df['title'] == title]
        for i in range(len(page_markers)):
            start_page = page_markers[i]
            end_page = page_markers[i+1] - 1 if i+1 < len(page_markers) else df_title['page_number'].max()
            mask = (df_title['page_number'] >= start_page) & (df_title['page_number'] <= end_page)
            sub_df = df_title.loc[mask]

            if sub_df.empty:
                continue

            merged_author = MergeAuthor(sub_df)

            first_row_of_range = sub_df.index[0]
            rows_to_keep.append(first_row_of_range)

            # 병합된 값 할당
            df.loc[first_row_of_range, 'author'] = merged_author
            for col in number_columns:
                df.loc[first_row_of_range, col] = MergeNums(sub_df, col)
            for col in string_columns:
                df.loc[first_row_of_range, col] = MergeStrs(sub_df, col)
            for col in list_columns_unique:
                df.loc[first_row_of_range, col] = str(MergeListsUnique(sub_df, col))
            for col in list_append_columns:
                df.loc[first_row_of_range, col] = str(MergeList(sub_df, col))
            df.loc[first_row_of_range, 'speaker_listener'] = str(MergeDicts(sub_df, 'speaker_listener'))

    df = df.loc[rows_to_keep]
    return df

df_merged = MergebyMarkers(df.copy(), markers)
df_merged
```
