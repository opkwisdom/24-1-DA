```python
import numpy as np

def group_identical_terms(words):
    grouped_terms = {}
    
    for word in words:
        added_to_group = False
        for key, group in grouped_terms.items():
            if key in word:
                grouped_terms[word] = [word]
                grouped_terms.pop(key)
                added_to_group = True
                break
            for term in group:
                if word in term:
                    grouped_terms[term].append(word)
                    added_to_group = True
                    break
            if added_to_group:
                break
        
        if not added_to_group:
            grouped_terms[word] = [word]
    
    return grouped_terms

ids = np.arange(37)
identical_terms = {}

for idx in ids:
    # 그룹화된 결과 출력
    grouped_terms = group_identical_terms(character_keys[keys[idx]])
    identical_terms[keys[idx]] = grouped_terms
    
```

```python
def replace_terms_with_keys(dictionary, terms_dict):
    replaced_dictionary = {}
    for key, values in dictionary.items():
        if isinstance(values, list):
            replaced_values = []
            for value in values:
                replaced_key = find_key_by_value(terms_dict, value)
                if replaced_key:
                    replaced_values.append(replaced_key)
                else:
                    replaced_values.append(value)
            replaced_dictionary[key] = replaced_values
        else:
            replaced_dictionary[key] = values
    return replaced_dictionary

# 예시 speaker_listener 딕셔너리
speaker_listener = {'화자': ["부모"], '청자': ["종"], "num_sentence": 5}

# speaker_listener 안의 값들을 identical_terms를 참조하여 대체
result = replace_terms_with_keys(speaker_listener, identical_terms['광작(廣作)'])

print(result)
```

## 화자, 청자 변수 생성하기

* 대사 바로 앞 문장 기준으로 변수 생성함!

1. 주격 조사 (보조사) + 부사격조사 ("JKS", "JX", "JKB") - O
2. 주격 조사 (보조사) ("JKS", "JX") - O
3. 부사격 조사 ("JKB") - O
4. 접속 조사가 같이 등장 - 판단 보류 ("JC") - O

```python
def find_speaker_listener(idx, cha_list):
    """
    idx: 데이터프레임 행 번호, cha_list: NER-tagged-set
    """
    punct_list = ["\"", "\'", ".", ",", ";", "!", "?"]
    
    # 마지막 문장 정제
    last_sentence = re.split(r'[.!?]', df.loc[idx, "quote_dup_fixed"])
    last_sentence = [l for l in last_sentence if l not in punct_list]
    last_sentence = [l for l in last_sentence if l]
    last_sentence = last_sentence[-1].lstrip()
    last_sentence = re.sub(r'[\"\'\.\,\;\!\?]', '', last_sentence)

    # print(i)
    # print(last_sentence)
    tagged_pos = tagger.pos(last_sentence)

    speaker = []
    listener = []
    temp_list = []

    for i, (label, tag) in enumerate(tagged_pos):
        if i > 0 and (tag == "JKS") or (tag == "JX"): # 주격 조사, 보조사
            if temp_list:
                for a in temp_list:
                    speaker.append(a)
                temp_list = []
            if i > 1 and tagged_pos[i-1][1] == "XSN":     # 명사 파생 접미사
                target = tagged_pos[i-2][0] + tagged_pos[i-1][0]
                speaker.append(target)
            else:
                speaker.append(tagged_pos[i-1][0])
        if i > 0 and tag == "JKB":  # 부사격 조사
            if temp_list:
                for a in temp_list:
                    listener.append(a)
                temp_list = []
            if i > 1 and tagged_pos[i-1][1] == "XSN":     # 명사 파생 접미사
                target = tagged_pos[i-2][0] + tagged_pos[i-1][0]
                listener.append(target)
            else:
                listener.append(tagged_pos[i-1][0])
        if i > 0 and tag == "JC":   # 접속 조사일 경우, 판단 보류
            temp_list.append(tagged_pos[i-1][0])

    speaker = [s for s in speaker if s in cha_list]
    listener = [l for l in listener if l in cha_list]

    if len(speaker) > 0:
        speaker = list(set(speaker))
    if len(listener) > 0:
        listener = list(set(listener))
    
    return {"화자": speaker, "청자": listener}
```

```python
def remove_duplicate_sentences(text, is_quote=True):
    # 따옴표 안의 문장들을 분리하여 리스트에 저장
    if is_quote:
        sentences_in_quotes = re.findall(r'\"(.*?)\"', text)
    else:
        # 문장 구분 기준으로 텍스트를 분리합니다.
        sentences_in_quotes = re.split(r'[.!?]', text)
        sentences_in_quotes = [s.strip() for s in sentences_in_quotes]


    unique_sentences = []
    seen_sentences = set()

    # 중복된 문장을 제거
    for sentence in sentences_in_quotes:
        if sentence not in seen_sentences:
            seen_sentences.add(sentence)
            unique_sentences.append(sentence)
    
    # 중복을 제거한 문장들을 다시 따옴표로 묶어서 반환
    if is_quote:
        res = ' '.join([f'"{sentence}"' for sentence in unique_sentences])
    else:
        res = ' '.join([sentence for sentence in unique_sentences])
    return res

print(remove_duplicate_sentences(df3.loc[200, "cleaned_content"]))
print(remove_duplicate_sentences("할미가 죽었으니 너는 낭자를 잘 보호하라. 할미가 죽었으니 너는 낭자를 잘 보호하라.", False))
```

```python
punct_list = [".", "?", "!"]

# 구두점을 기준으로 문장을 분할하여 리스트로 반환하는 함수
def split_sentences(text):
    for punct in punct_list:
        text = text.replace(punct, punct + "|")  # 구두점을 기준으로 분할하기 위해 구두점 뒤에 "|" 추가
    sentences = text.split("|")
    sentences = [sentence.strip() for sentence in sentences if sentence.strip()]  # 공백을 제거하고 빈 문장 제거
    return sentences
```

```python
def check_sentences_in_quotes(text):
    # 쌍따옴표 안에 있는 문장 추출
    sentences_in_quotes = re.findall(r'"([^"]*)"', text)
    
    # 쌍따옴표 안에 있는 문장이 있는지 확인
    if sentences_in_quotes:
        # 쌍따옴표 안에 있는 문장이 있는 경우
        print("문장이 쌍따옴표 안에 있습니다.")
        print("쌍따옴표 안에 있는 문장:", sentences_in_quotes)
        
        # 쌍따옴표 안에 있는 문장만으로 이루어진 텍스트 생성
        text_in_quotes = " ".join(sentences_in_quotes)
        
        # 원본 텍스트와 쌍따옴표 안에 있는 문장만으로 이루어진 텍스트가 일치하는지 확인
        if text == text_in_quotes:
            return True
        else:
            return False
    else:
        # 쌍따옴표 안에 있는 문장이 없는 경우
        print("문장이 쌍따옴표 안에 없습니다.")
        return False
    
check_sentences_in_quotes(df.loc[12945, "quote_dup"])
```
