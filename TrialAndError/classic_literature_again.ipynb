# 각주 전처리

```python
df_desc = df[~df["text_desc"].isna()].reset_index(drop=True)
df_desc
```

```python
df_desc["text_desc_split"] = df_desc["text_desc"].apply(lambda x: x.split("\n"))
df_desc
```

```python
# 각주 문장별 split
df_desc_explode = df_desc.explode(column="text_desc_split").reset_index(drop=True)
df_desc_explode = df_desc_explode[~(df_desc_explode["text_desc_split"] == '')].reset_index(drop=True)
df_desc_explode
```

```python
# 함수 정의
def safe_split(x, index):
    try:
        return x.split(':')[index]
    except IndexError:
        return None

df_desc_explode["annot"] = df_desc_explode["text_desc_split"].apply(safe_split, index=0)
df_desc_explode["annot_desc"] = df_desc_explode["text_desc_split"].apply(safe_split, index=1)
df_desc_explode
```

```python
df_desc_explode = df_desc_explode[~df_desc_explode["annot_desc"].isna()].reset_index(drop=True)
df_desc_explode
```

# Annot 전처리

```python
test = df_desc_explode["annot"][0]
test
```

```python
import re


pattern = re.compile(r'\[.*\] ')
re.sub(pattern, '', test)
```

```python
# [] 안에 있는 말 제거
pattern = re.compile(r'\[.*\] ')

df_desc_explode["annot"] = df_desc_explode["annot"].apply(lambda x: re.sub(pattern, '', x))
df_desc_explode
```

```python
# () 패턴 제거
pattern = re.compile(r'\(.*\)')

df_desc_explode["annot"] = df_desc_explode["annot"].apply(lambda x: re.sub(pattern, '', x))
df_desc_explode
```

```python
# 마지막 공백 제거
pattern = re.compile(r'\s+$')

df_desc_explode["annot"] = df_desc_explode["annot"].apply(lambda x: re.sub(pattern, '', x))
df_desc_explode
```

```python
df_desc_explode.to_csv("classic_annot.csv", index=False)
```
