re 연산을 통해 한글 & 일반적인 특수문자(쉼표, 따옴표, 괄호 등)을 제외한 모든 문자를 삭제

```python
import pandas as pd
import re

def clean_text(text):
    # 한글과 특수문자 외의 모든 문자를 지우는 정규표현식
    #     한글 유니코드 범위: 완성형 한글 (U+AC00-U+D7AF), 한글 자모 (U+1100-U+11FF)
    #     특수문자 범위: 쉼표, 마침표, 느낌표, 물음표, 세미콜론, 콜론, 따옴표, 괄호 등
    pattern = '[^\uAC00-\uD7AF\u1100-\u11FF0-9\s,\.!?;:\(\)\[\]\'\"“”‘’~-]'

    # 텍스트에 적용
    cleaned_text = re.sub(pattern, '', str(text))
    return cleaned_text

df['cleaned_content'] = df['content'].apply(clean_text)
df[['content', 'cleaned_content']]
```
